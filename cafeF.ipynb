```python
# lấy data của 5 link đầu trong phần kế hoạch đầu tư

driver.get(href_link_1)

# Tìm tất cả các phần tử div có class "col-sm-24 col-md-14"
div_elements = driver.find_elements(By.CSS_SELECTOR,'div.col-sm-24.col-md-14')

# Lọc ra 5 liên kết thỏa mãn điều kiện
filtered_links = []
for div in div_elements:
    link_elements = div.find_elements(By.CSS_SELECTOR,'a.text-link.news-link')
    for link_element in link_elements:
        href = link_element.get_attribute('href')
        if href:
            filtered_links.append(href)
            if len(filtered_links) == 5:
                break
    if len(filtered_links) == 5:
        break

# In ra 5 liên kết đã lọc
for link in filtered_links:
    print(link)
```

```python
from googlesearch import search
import requests  # Import thư viện requests

root = ['cafef', 'finance.vietstock', 'vnexpress', 'vn.investing.com']

# Tìm kiếm
query = input()
search_results = []

# User-Agent giả mạo để tránh lỗi 429
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.36"
}

# Đặt tùy chọn headers trong yêu cầu
for j in search(query, tld="com.vn", num=10, stop=10, pause=2):
    # Tạo yêu cầu với User-Agent
    response = requests.get(j, headers=headers)
    
    # Kiểm tra xem yêu cầu thành công
    if response.status_code == 200:
        search_results.append(j)

# Kiểm tra và in ra các kết quả chứa từ trong danh sách root
for result in search_results:
    for keyword in root:
        if keyword in result:
            print(result)
            # break
```
